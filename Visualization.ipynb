```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic data with correlated features
np.random.seed(42)
size = 100
x = np.random.normal(0, 1, size)
y = 2 * x + np.random.normal(0, 1, size)  # Correlated with x
z = 0.5 * x + np.random.normal(0, 1, size)  # Less correlated with x

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Compute the correlation matrix
corr_matrix = data.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap', fontsize=14)
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.

# np.random.seed(20)
# data = {
#     'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#     'Day': np.random.choice(range(1, 31), 100),
#     'Sales': np.random.randint(1000, 5000, 100)
# }
# df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.0f', linewidths=0.5)

# Add title and labels
plt.title('Sales Variation Across Different Months and Days', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Day', fontsize=12)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none', alpha=0.8)

# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Add title and labels
ax.set_title('3D Surface Plot', fontsize=14)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'], alpha=0.7, color='skyblue', edgecolor='black')

# Annotate countries
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i] + 20, df['Population'][i], country, fontsize=10)

# Add title and labels
plt.title('Bubble Chart: Population vs GDP', fontsize=14)
plt.xlabel('GDP (in billions)', fontsize=12)
plt.ylabel('Population (in millions)', fontsize=12)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']â€¨
# counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge
from bokeh.io import output_notebook

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure
p = figure(x_range=fruits, title="Fruit Counts", width=800, height=400,
           toolbar_location=None, tools="")

# Add a bar chart
p.vbar(x='fruits', top='counts', width=0.6, source=source, color="skyblue", legend_label="Count")

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.visible = False
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"

# Display the chart
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data
# data_hist = np.random.randn(1000)â€¨
# hist, edges = np.histogram(data_hist, bins=30)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate data and compute histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],
    right=edges[1:]
))

# Create a figure
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency",
           width=800, height=400)

# Add quad glyphs to create the histogram
p.quad(top="top", bottom=0, left="left", right="right", source=source,
       fill_color="blue", line_color="black", alpha=0.7)

# Display the plot
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset
# data_heatmap = np.random.rand(10, 10)â€¨
# x = np.linspace(0, 1, 10)â€¨
# y = np.linspace(0, 1, 10)â€¨
# xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
import numpy as np

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Create a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create a figure
p = figure(title="Bokeh Heatmap", x_axis_label="X", y_axis_label="Y",
           width=800, height=400, toolbar_location=None, tools="")

# Add the heatmap
p.rect(x="x", y="y", width=0.1, height=0.1, source=source,
       fill_color={'field': 'value', 'transform': color_mapper}, line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0), title="Value")
p.add_layout(color_bar, 'right')

# Display the heatmap
show(p)
```
